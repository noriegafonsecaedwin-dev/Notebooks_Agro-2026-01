## Tres Ejemplos de Código (Sencillos y Útiles)

Aquí tienes tres aplicaciones prácticas que podrías empezar a usar hoy mismo:

### 1. Cálculo del Índice de Madurez

En agroindustria, determinar el momento exacto de cosecha es clave. Este programa calcula la relación entre sólidos solubles ($^{\circ}Brix$) y la acidez.

```
# Cálculo simple de Índice de Madurez
grados_brix = 12.5
acidez_titulable = 0.8

indice_madurez = grados_brix / acidez_titulable

print(f"El índice de madurez es: {indice_madurez:.2f}")

if indice_madurez > 15:
    print("Estado: Óptimo para procesamiento.")
else:
    print("Estado: Aún requiere tiempo de maduración.")
```

### 2. Control de Humedad en Secado

Un script básico para monitorear si un producto (como granos de café o cereales) ha alcanzado el porcentaje de humedad deseado para su almacenamiento.

```
humedad_actual = 18.5
limite_seguro = 13.0

while humedad_actual > limite_seguro:
    print(f"Humedad actual: {humedad_actual}% - Continuando secado...")
    # Simulamos que baja la humedad
    humedad_actual -= 1.5

print("Humedad alcanzada. ¡Apagando secadoras!")
```

### 3. Estimación de Rendimiento de Extracción

Ideal para procesos donde se extraen aceites o jugos, permitiendo evaluar la eficiencia de la maquinaria.

```
def calcular_rendimiento(masa_inicial, masa_final):
    rendimiento = (masa_final / masa_inicial) * 100
    return rendimiento

# Ejemplo: Extracción de aceite de palma
fruta_ingresada = 500  # kg
aceite_obtenido = 110  # kg

resultado = calcular_rendimiento(fruta_ingresada, aceite_obtenido)
print(f"El rendimiento de la extracción fue del {resultado}%")
```
