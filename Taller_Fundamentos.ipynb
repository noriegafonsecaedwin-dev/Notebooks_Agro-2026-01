### Ejercicio 1: Capacidad de Silos de Granos

**Planteamiento del Problema:**
Se requiere calcular cuántas toneladas de maíz puede almacenar un silo cilíndrico. El usuario ingresa el radio y la altura del silo en metros.

**Requerimientos del problema:**
*   Solicitar al usuario el radio (en metros) del silo.
*   Solicitar al usuario la altura (en metros) del silo.
*   Utilizar la constante π (pi).
*   Aplicar la fórmula del volumen de un cilindro: `V = π × r² × h`.
*   Utilizar la densidad del maíz: `0.75 ton/m³`.
*   Calcular la capacidad total en toneladas.
*   Mostrar el resultado de la capacidad en toneladas.

**Prueba del programa:**
Radio = 3m, Altura = 10m. Salida esperada ≈ 212.06 toneladas.

```python
import math

# Solicitar datos al usuario
radio = float(input("Ingrese el radio del silo en metros: "))
altura = float(input("Ingrese la altura del silo en metros: "))

# Datos
densidad_maiz = 0.75 # ton/m3

# Calcular el volumen del silo
volumen = math.pi * (radio ** 2) * altura

# Calcular la capacidad en toneladas
capacidad = volumen * densidad_maiz

# Mostrar el resultado
print(f"La capacidad del silo es de {capacidad:.2f} toneladas.")
```

### Explicación del Código (Ejercicio 1)

*   `import math`: Importa el módulo `math` para poder usar la constante `math.pi` (el valor de pi).
*   `radio = float(input("..."))`: Solicita al usuario que ingrese el radio del silo y convierte la entrada (que es una cadena de texto) a un número de punto flotante (`float`).
*   `altura = float(input("..."))`: Solicita la altura del silo y la convierte a `float`.
*   `densidad_maiz = 0.75`: Define la densidad del maíz en toneladas por metro cúbico.
*   `volumen = math.pi * (radio ** 2) * altura`: Calcula el volumen del silo utilizando la fórmula `π × r² × h`.
*   `capacidad = volumen * densidad_maiz`: Calcula la capacidad total en toneladas multiplicando el volumen por la densidad del maíz.
*   `print(f"...")`: Imprime el resultado formateado a dos decimales (`:.2f`) para mayor claridad.

### Ejercicio 2: Dosificación de Fertilizante Líquido

**Planteamiento del Problema:**
Un sistema de riego por goteo requiere aplicar una dosis de fertilizante por hectárea. El usuario ingresa las hectáreas totales y la dosis requerida (litros/hectárea).

**Requerimientos del problema:**
*   Solicitar al usuario las hectáreas totales.
*   Solicitar al usuario la dosis requerida de fertilizante por hectárea (litros/hectárea).
*   Calcular el total de fertilizante puro necesario.
*   Aplicar la relación de dilución 1:500 (1 litro de producto por 500 de agua).
*   Calcular la cantidad total de agua necesaria.
*   Mostrar la cantidad de producto y la cantidad de agua total.

**Prueba del programa:**
5 hectáreas, Dosis 2 L/ha. Salida esperada: 10L de producto y 5000L de agua total.

```python
# Solicitar datos al usuario
hectareas_totales = float(input("Ingrese las hectáreas totales: "))
dosis_por_hectarea = float(input("Ingrese la dosis requerida de fertilizante (litros/hectárea): "))

# Datos
relacion_dilucion_agua = 500 # 1 litro de producto por 500 de agua

# Calcular la cantidad total de fertilizante puro (producto)
litros_producto = hectareas_totales * dosis_por_hectarea

# Calcular la cantidad total de agua necesaria
litros_agua = litros_producto * relacion_dilucion_agua

# Mostrar los resultados
print(f"Se necesitan {litros_producto:.2f} L de producto de fertilizante.")
print(f"Se necesitan {litros_agua:.2f} L de agua total para la dilución.")
```

### Explicación del Código (Ejercicio 2)

*   `hectareas_totales = float(input("..."))`: Pide al usuario el número total de hectáreas y lo convierte a `float`.
*   `dosis_por_hectarea = float(input("..."))`: Pide la dosis de fertilizante por hectárea y lo convierte a `float`.
*   `relacion_dilucion_agua = 500`: Establece la constante de la relación de dilución (por cada litro de producto, 500 litros de agua).
*   `litros_producto = hectareas_totales * dosis_por_hectarea`: Calcula la cantidad total de fertilizante puro (producto) multiplicando las hectáreas por la dosis por hectárea.
*   `litros_agua = litros_producto * relacion_dilucion_agua`: Calcula la cantidad total de agua necesaria multiplicando el fertilizante puro por la relación de dilución.
*   `print(f"...")`: Muestra la cantidad de producto y de agua, formateando los resultados a dos decimales.

### Ejercicio 3: Rendimiento de Pulpa en Frutas

**Planteamiento del Problema:**
Una planta procesadora necesita saber el porcentaje de aprovechamiento de la fruta. Se ingresa el peso total de la fruta comprada y el peso de los residuos (cáscaras y semillas).

**Requerimientos del problema:**
*   Solicitar al usuario el peso total de la fruta (en kg).
*   Solicitar al usuario el peso de los residuos (en kg).
*   Calcular el peso de la pulpa: `Pulpa = Peso Total - Residuos`.
*   Calcular el porcentaje de rendimiento: `%Rendimiento = (Pulpa / Peso Total) × 100`.
*   Mostrar el peso de la pulpa y el porcentaje de rendimiento.

**Prueba del programa:**
Peso Total = 500kg, Residuos = 150kg. Salida esperada: 350kg de pulpa (70%).

```python
# Solicitar datos al usuario
peso_total = float(input("Ingrese el peso total de la fruta comprada (kg): "))
peso_residuos = float(input("Ingrese el peso de los residuos (kg): "))

# Calcular el peso de la pulpa
peso_pulpa = peso_total - peso_residuos

# Calcular el porcentaje de rendimiento
porcentaje_rendimiento = (peso_pulpa / peso_total) * 100

# Mostrar los resultados
print(f"El peso de la pulpa es de {peso_pulpa:.2f} kg.")
print(f"El porcentaje de rendimiento es del {porcentaje_rendimiento:.2f} %.")
```

### Explicación del Código (Ejercicio 3)

*   `peso_total = float(input("..."))`: Pide al usuario el peso total de la fruta y lo convierte a `float`.
*   `peso_residuos = float(input("..."))`: Pide el peso de los residuos y lo convierte a `float`.
*   `peso_pulpa = peso_total - peso_residuos`: Calcula el peso de la pulpa restando los residuos del peso total.
*   `porcentaje_rendimiento = (peso_pulpa / peso_total) * 100`: Calcula el porcentaje de rendimiento de la pulpa respecto al peso total de la fruta.
*   `print(f"...")`: Muestra el peso de la pulpa y el porcentaje de rendimiento, formateando los resultados a dos decimales.

### Ejercicio 4: Dosificación de Mortero para Muros

**Planteamiento del Problema:**
Calcular cuántos bultos de cemento se necesitan para un muro de ladrillo. El usuario ingresa el área del muro en m².

**Requerimientos del problema:**
*   Solicitar al usuario el área del muro (en m²).
*   Utilizar el dato de que `0.25 bultos de cemento` se requieren por cada `1 m² de muro`.
*   Calcular el número total de bultos de cemento necesarios.
*   Mostrar el resultado de la cantidad de bultos.

**Prueba del programa:**
Muro de 40 m². Salida esperada: 10 bultos de cemento.

```python
# Solicitar datos al usuario
area_muro = float(input("Ingrese el área del muro en m²: "))

# Datos
bultos_por_m2 = 0.25 # bultos de cemento por m2

# Calcular la cantidad de bultos de cemento
bultos_cemento = area_muro * bultos_por_m2

# Mostrar el resultado
print(f"Se necesitan {bultos_cemento:.2f} bultos de cemento para el muro.")
```

*   `print(f"...")`: Muestra el número de bultos de cemento, formateando el resultado a dos decimales.

### Ejercicio 5: Peso de una Viga de Acero

**Planteamiento del Problema:**
Un ingeniero necesita calcular el peso propio de una viga de sección rectangular para el análisis estructural. Se ingresa base, altura y longitud en metros.

**Requerimientos del problema:**
*   Solicitar al usuario la base de la viga (en metros).
*   Solicitar al usuario la altura de la viga (en metros).
*   Solicitar al usuario la longitud de la viga (en metros).
*   Utilizar la densidad del acero: `7850 kg/m³`.
*   Aplicar la fórmula del peso: `Peso = Base × Altura × Longitud × Densidad`.
*   Mostrar el resultado del peso en kilogramos.

**Prueba del programa:**
Base = 0.3m, Altura = 0.5m, Largo = 6m. Salida esperada: 7065 kg.

```python
# Solicitar datos al usuario
base = float(input("Ingrese la base de la viga en metros: "))
altura_viga = float(input("Ingrese la altura de la viga en metros: "))
longitud = float(input("Ingrese la longitud de la viga en metros: "))

# Datos
densidad_acero = 7850 # kg/m3

# Calcular el peso de la viga
peso_viga = base * altura_viga * longitud * densidad_acero

# Mostrar el resultado
print(f"El peso de la viga de acero es de {peso_viga:.2f} kg.")
```

### Explicación del Código (Ejercicio 5)

*   `base = float(input("..."))`: Pide al usuario la base de la viga y la convierte a `float`.
*   `altura_viga = float(input("..."))`: Pide la altura de la viga y la convierte a `float`.
*   `longitud = float(input("..."))`: Pide la longitud de la viga y la convierte a `float`.
*   `densidad_acero = 7850`: Define la densidad del acero en kilogramos por metro cúbico.
*   `peso_viga = base * altura_viga * longitud * densidad_acero`: Calcula el peso de la viga multiplicando sus dimensiones por la densidad del acero.
*   `print(f"...")`: Muestra el peso de la viga, formateando el resultado a dos decimales.

### Ejercicio 6: Pendiente de una Rampa de Acceso

**Planteamiento del Problema:**
Determinar si una rampa cumple con la normativa de accesibilidad. El usuario ingresa la altura a vencer y la longitud horizontal de la rampa (en metros).

**Requerimientos del problema:**
*   Solicitar al usuario la altura a vencer (en metros).
*   Solicitar al usuario la longitud horizontal de la rampa (en metros).
*   Aplicar la fórmula del porcentaje de pendiente: `%Pendiente = (Altura / Longitud) × 100`.
*   Mostrar el resultado del porcentaje de pendiente.

**Prueba del programa:**
Altura = 0.6m, Longitud = 5m. Salida esperada: 12% de pendiente.

```python
# Solicitar datos al usuario
altura_rampa = float(input("Ingrese la altura a vencer de la rampa en metros: "))
longitud_rampa = float(input("Ingrese la longitud horizontal de la rampa en metros: "))

# Calcular el porcentaje de pendiente
porcentaje_pendiente = (altura_rampa / longitud_rampa) * 100

# Mostrar el resultado
print(f"La pendiente de la rampa es del {porcentaje_pendiente:.2f} %.")
```

### Explicación del Código (Ejercicio 6)

*   `altura_rampa = float(input("..."))`: Pide al usuario la altura a vencer de la rampa y la convierte a `float`.
*   `longitud_rampa = float(input("..."))`: Pide la longitud horizontal de la rampa y la convierte a `float`.
*   `porcentaje_pendiente = (altura_rampa / longitud_rampa) * 100`: Calcula el porcentaje de pendiente dividiendo la altura por la longitud y multiplicando por 100.
*   `print(f"...")`: Muestra el porcentaje de pendiente, formateando el resultado a dos decimales.
